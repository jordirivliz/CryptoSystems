### Problem 1: Flow control

Write a function named **fizz_buzz** that takes as input an integer **N** and for each integer **m** from 1 to **N**, and does two things:
 - prints 'Fizz' if **m** is divisible by 2 and not 5, prints 'Buzz' if **m** is divisible by 5 and not 2, prints 'FizzBuzz' if **m** is divisible by 2 and 5, prints 'Moot' if none of the above;
 - returns as output a list of these print statements

eg.
Calling fizz_buzz(10)

should produce
```
Moot
Fizz
Moot
Fizz
Buzz
Fizz
Moot
Fizz
Moot
FizzBuzz
['Moot', 'Fizz', 'Moot', 'Fizz', 'Buzz', 'Fizz', 'Moot', 'Fizz', 'Moot', 'FizzBuzz']
```

````python
#1 Solution
def fizz_buzz(N):
    output = []
    if type(N) != int:
        output.append('N is not an int')
        return output
    for m in range(1, N+1):
        if m%2 == 0 and m%5 != 0:
            print('Fizz')
            output.append('Fizz')
        elif m%2 != 0 and m%5 == 0:
            print('Buzz')
            output.append('Buzz')
        elif m%2 == 0 and m%5 == 0:
            print('FizzBuzz')
            output.append('FizzBuzz')
        else:
            print('Moot')
            output.append('Moot')
    return output
````

a. Write a function **str2num(s)** that takes as input a string of characters, capitalizes it and returns the SAGE integer obtained by concatenating the corresponding  ASCII codes, e.g. str2num("""A cat!""") returns $653267658433$ (a SAGE integer, not a string and not a Python integer!).
````python
# 2a Solution:
def str2num(s):
    s = s.upper()
    output = ''
    for char in s:
        num = ord(char)
        output = output + str(num)
    return int(output)
````

````python
# 2c Solution:
import string
def remove_punctuation(s):
    output = ''
    if type(s) != str:
        output = 'Invalid input'
        return output
    s = s.upper()
    alphabet = string.ascii_uppercase
    l = list(alphabet)
    l.extend(['.', ',', '!', '?', ' ', '"', ':', ';', '...', '-', '_', '–', '(', ')', '[', ']', '\''])
    for char in s:
        if char not in l:
            return 'Invalid input'
        else:
            if char.isalnum():
                output = output + char
    return output
````

````python
# 2d Solution:
def str26(s):
    output = []
    t = remove_punctuation(s)
    if t == 'Invalid input':
        return 'Invalid input'
    for char in t:
        num = ord(char)
        num = num - 65
        output.append(num)
    return output
````

### Problem 3: Caesar shifts

Again, some of this is very similar to a problem from HRead about Caesar shifts at https://en.wikipedia.org/wiki/Caesar_cipher


a. Write a function **ngram(s,N)** that takes as input a string **s**, and a positive integer **N**; returns the string **s** separated into N-grams, all in capital letters and stripped of punctuation. For instace ngram('A cat!', 3) should return 'ACA T'. If **N** is not a positive integer or **s** is not a string, or if s contains numbers or special characters, it should return the message 'Invalid input'.



b. Write a function **encrypt_caesar_gram (s,k,N)** that takes as input a string **s**, a key **k** that consists of a single letter, and a positive integer **N**; returns the **k** Caesar shift of the string **s** stripped of spaces and punctuation into **N**-grams and written as upper case letters. If **s** is not a string, if **s** contains special characters, if **k** is not a single letter or if **N** is not a positive integer, it should return the message 'Invalid input'. That is, if you could not do this by hand, then your function should return the error message. 


c. Write a function **decrypt_caesar(s,k)** that takes as input a string **s**, and a key **k** that consists of a single letter; applies **remove_punctuation(s)**, and then decrypts the Caesar shift given by **k** on the stripped string **remove_punctuation(s)**. The output should be a string with no spaces and containing only upper case letters. If the string **s** is not of the expected form (e.g. it contains digits or special characters), it should return the error message 'Invalid input'. If the key **k** is anything other than a single letter, it should return the error message 'Invalid input'. (You might find the functions from Problem 1 useful.)

````python
# 3a Solution:
def ngram(s,N):
    if N <= 0:
        return 'Invalid input'
    t = remove_punctuation(s)
    if t == 'Invalid input':
        return 'Invalid input'
    t = ' '.join(t[i: i+N] for i in range(0, len(t), N))
    return t
````

````python
# 3b Solution:
def encrypt_caesar_gram(s,k,N):
    s = ngram(s,N)
    if s == 'Invalid input':
        return 'Invalid input'
    if len(k) > 1 or len(k) == 0:
        return 'Invalid input'
    original = str26(s)
    t = str26(k)
    if t == 'Invalid input':
        return 'Invalid input'
    leng = len(original)
    encrypt = [None]* leng
    for i in range(0, leng):
        num = (original[i] + t[0])%26
        encrypt[i] = num
    output = ''
    for val in encrypt:
        asc = chr(val+65)
        output = output + asc
    output = ngram(output, N)
    return output
````

````python
# 3c Solution:
def decrypt_caesar(s,k):
    s = remove_punctuation(s)
    if s == 'Invalid input':
        return 'Invalid input'
    s = s.upper()
    if len(k) > 1 or len(k) == 0:
        return 'Invalid input'
    if s == 'Invalid input':
        return 'Invalid input'
    t = str26(k)
    if t == 'Invalid input':
        return 'Invalid input'
    original = str26(s)
    leng = len(original)
    decrypt = [None]* leng
    for i in range(0, leng):
        num = (original[i] - t[0])%26
        decrypt[i] = num
    output = ''
    for val in decrypt:
        asc = chr(val+65)
        output = output + asc
    return output
````

### Problem 4: Vigenere

Read about the Vigenere cipher at https://en.wikipedia.org/wiki/Vigen%C3%A8re_cipher
We will only use the concepts under **Description**.

a. Write a function **encrypt_vigenere (s, k, N)** that takes as input two strings **s** and **k** and a positive integer **N**; removes the spaces and punctuation from both of them; and then encrypts the stripped string **s** by a Vigenere cipher with the key given by the stripped string **k** written as **N**-grams of capital letters. If either **s** or **k** is not a string or contain digits or special characters, or if **k** is the empty string, it should return the message 'Invalid input'. Same if **N** is not a positive integer.

b. Write a function **decrypt_vigenere (s, k)** that takes as input two strings **s** and **k**; removes the spaces and punctuation from both of them; and then performs the Vigenere decryption of the stripped string **s** assuming the encrytion key used was the stripped string **k**. The output should be a string consisting only of capital letters. If either **s** or **k** contain digits or special characters, it should return the message 'Invalid input'. Same if **N** is not a positive integer.

````python
# 4a Solution:
def encrypt_vigenere(s,k,N):
    s = remove_punctuation(s)
    if s == 'Invalid input':
        return 'Invalid input'
    k = remove_punctuation(k)
    if k == 'Invalid input':
        return 'Invalid input'
    if N <= 0:
        return 'Invalid input'
    t = str26(s)
    if t == 'Invalid input': return t
    if len(k) == 0: return 'Invalid input'
    if str26(k) == 'Invalid input': return 'Invalid input'
    key = (str26(k)*len(s))[:len(s)]
    output = ''
    for i in range(len(t)):
        output += chr(((t[i]+key[i])%26)+65)
    output = ngram(output, N)
    return output
````

## Problem 5 (Bonus): Fibonacci numbers

In this problem you have to use the recurrence formula for Fibonacci numbers $f_n = f_{n-1} + f_{n-2}, \quad f_1 = f_2 =1.$

a. Write an _efficient_ function **fib(N)** that takes input a positive integer **N** and returns the Nth Fibonacci number.

b. Now use this to write another function **allfib(N)** that takes input a positive integer **N** and returns a list of the first N Fibonacci numbers.

Warning: your code should be efficient enough so that all tests take less than 30 seconds.
Hint: it might be easier to solve the second part first.

````python
#5a Solution

# YOUR CODE HERE
def fib(N):
    list = [None] * (N+2)
    list[0] = 1
    list[1] = 1
    for i in range(2, N):
        list[i] = list[i-1] + list[i-2]
    return list[N-1]
````

````python
#5b Solution
def allfib(N):
    if N == 1:
        return [1]
    list = [None] * (N+2)
    list[0] = 1
    list[1] = 1
    for i in range(2, N):
        list[i] = list[i-1] + list[i-2]
    result = []
    for num in list:
        if num != None:
            result.append(num)
    return result
````
